```python
from utils import *
from sklearn.model_selection import GridSearchCV
```

```python
chips_path = './data/chips.csv'
geyser_path = './data/geyser.csv'
jobs_count = 11
cv_count = 5
test_size = 0.3
```

```python
X_chips, Y_chips = read_data(chips_path)
X_geyser, Y_geyser = read_data(geyser_path)
```

```python
parameters_acc = {
    'trees_cnt': [1, 10, 100, 200, 500, 1000]
}
clf_chips_acc = GridSearchCV(DecisionForest(), parameters_acc, cv=cv_count, n_jobs=jobs_count, scoring='accuracy')
clf_chips_acc.fit(X_chips, Y_chips)
```

```python
print(clf_chips_acc.best_params_)
print(clf_chips_acc.best_score_)
```

```python
clf_geyser_acc = GridSearchCV(DecisionForest(), parameters_acc, cv=cv_count, n_jobs=jobs_count, scoring='accuracy')
clf_geyser_acc.fit(X_geyser, Y_geyser)
```

```python
print(clf_geyser_acc.best_params_)
print(clf_geyser_acc.best_score_)
```
